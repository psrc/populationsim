```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats
import os

%matplotlib inline

pd.options.display.float_format = '{:0,.0f}'.format
```

```python
cwd = os.getcwd()
data_path = os.path.join(os.path.dirname(cwd), 'output', 'summary_adjusted.csv')  
df = pd.read_csv(data_path)
```

```python
def plot_chart(df, field_list, xfield='_control', yfield='_adjusted'):
    for field in field_list:
        ax = df.plot(kind='scatter', x=field+xfield, y=field+yfield)
        lim = df[[field+xfield,field+yfield]].max().max()*1.05
        ax.set_ylim(0,lim)
        ax.set_xlim(0,lim)
        x = np.linspace(*ax.get_xlim())
        ax.plot(x, x)
        slope, intercept, r_value, p_value, std_err = stats.linregress(df[field+xfield],df[field+yfield])
        print('r^2 ' + str(field) + ': ' + str(r_value))
        
        
    sum_df = pd.DataFrame()
    for source in ['control','adjusted','result']:
        df[field+'_'+source] = df[field+'_'+source].astype('float')
        sum_df[source] = df[[field+'_'+source for field in field_list]].sum().values
    sum_df.index = field_list
    display(sum_df)
```

# HH Size and Populations

# CrossTabs

```python
# Across household size and income

field_list_x =['income_lt15','income_gt15-lt30','income_gt30-lt60','income_gt60-lt100','income_gt100']
field_list_y = ['hh_size_1','hh_size_2','hh_size_3','hh_size_4','hh_size_5','hh_size_6','hh_size_7_plus']

_df = df.copy()
field_list = []
for source in ['_control','_adjusted']:
    for x in field_list_x:
        for y in field_list_y:
            _df[x+'_'+y+source] = _df[x+source]+_df[y+source]
    
        
            field_list.append(x+'_'+y)
```

```python
plot_chart(_df, field_list=field_list)
```
